#### 5. Rename the variable "Pos" to "Position". 

```python
nba_copy = nba.copy() # Make deep copy so orig dataset is not edited 
nba_copy.rename(columns = {"Pos":"Position"}, inplace = True) # inplace: If True then the data is modified directly, rather than returning shallow copy
nba_copy.head() 
```

#### 6. Use this variable to create two variables that are called "first_position" and "second_position".

```python
nba_copy[['first_position', 'second_position']] = nba_copy['Position'].str.split(pat = "-", expand=True)
nba_copy.head()[['Year', 'Player', 'first_position', 'second_position']]
```

#### 7. Combine these two variables back into a single variable called "position_united".

```python
nba_copy['position_united'] = np.where( nba_copy["second_position"].isnull(), nba_copy['first_position'], nba_copy['first_position'] + "-" + nba_copy['second_position'])
nba_copy.head()[['Year', 'Player', 'position_united', 'first_position', 'second_position']]
# nba_copy['position_united'].value_counts()  # Checking data via frequency tables
```

#### 8. Create two new datasets. <a class="anchor" id="bullet8"></a>

* a new dataset from the original dataset that includes all data except the age variable (be sure to give this dataset a new name).
* a new dataset from the original dataset that includes the year, the player name, and age.
* add a new column to both datasets called mergeid that includes a sequence of numbers beginning with a 1 in the first row of the data and ending with the total number of rows in the last row of the data


```python
nba_1 = nba.drop(columns=['Age']).copy()      # dataset with all except age
nba_2 = nba[['Year', 'Player', 'Age']].copy() # dataset with only year, name, and age

# Since both datasets are in the same order, we can create matching keys by using a numbered list of integers
nba_1['mergeid'] = range(1, len(nba_1) + 1) # add mergeid based on the length of the dataset
nba_2['mergeid'] = range(1, len(nba_2) + 1) # add mergeid based on the length of the dataset
```

```python
# Check data
print(nba_1.shape)
print(nba_1.columns)
print(nba_2.shape)
print(nba_2.columns)
```

#### 9. Join the two datasets from question (8) together to recreate the original dataset plus the new merge id. <a class="anchor" id="bullet9"></a>

```python
# Join the two datasets, you could also use a left or right join since the number of rows and unique keys match perfectly 
nba_merge = nba_1.merge(nba_2, how = 'inner', left_on = ['mergeid'], right_on = ['mergeid'])
nba_merge
```

#### 10. Subset the original dataset to the years 1995-1997. Group the data by team name and year then summarize the average number of points per team. Arrange from most to least points. <a class="anchor" id="bullet10"></a>
