## Project Overview
<a class="anchor" id="bullet2"></a>

The project I chose is **HowToCook**, a very popular Chinese-language GitHub repository.

GitHub project linhttps://github.com/Anduin2017/HowToCook?tab=readme-ov-fileToCook

This repository collects a large number of Chinese home-style recipes and presents them in a clear, structured way. It has gained significant attention on GitHub and is widely shared among Chinese-speaking users.


### Why this project is interesting
<a class="anchor" id="bullet3"></a>

I find this project interesting because it shows that GitHub is not only used for *software engineering*, but also for sharing **everyday knowledge**.

The project demonstrates how open-source platforms can be used creatively beyond traditional programming tasks.


## What this repository contains
<a class="anchor" id="bullet4"></a>

This repository includes:
- Step-by-step Chinese cooking instructions
- Images illustrating finished dishes
- Contributions from many GitHub users
- Clear organization using Markdown files
