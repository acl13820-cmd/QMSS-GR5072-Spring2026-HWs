### 1.[`10 pts`] Web-scrape all table data from the following [webpage](http://vincentarelbundock.github.io/Rdatasets/datasets.html) and build a pd.DataFrame object. Limit your final table to include columns for "Package", "Item", "Title", "Rows", and "Cols". Print the dimensions and first five rows of the table.

```python
# Send a GET request to the website and retrieve the HTML content
url = "https://vincentarelbundock.github.io/Rdatasets/datasets.html"
response = requests.get(url)

# Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(response.content, "html.parser")

# Find the first table in the HTML content
table = soup.find_all('table')[0]

# Read the table into a pandas DataFrame
df = pd.read_html(str(table))[0]

# Select required columns
df_col = df[["Package", "Item", "Title", "Rows", "Cols"]]

print(df_col.shape)
```

### 2.[`10 pts`] Web-scrape the full links to every CSV file listed in the CSV column of the web-page. Add a new column to your data frame that includes these links. Name the column "`csv_links`". Print the first five rows of the table.

```python
# define a function to extract html document from given url
def getHTMLdocument(url):
    # request for HTML document of given url
    response = requests.get(url)
    # response will be provided in JSON format
    return response.text

# create document
html_document = getHTMLdocument(url)
  
# create soup object
soup = BeautifulSoup(html_document, 'html.parser')

csv_list = []
for link in soup.find_all('a', attrs={'href': re.compile("^https://")}):
    href = link.get('href')
    
    if href.endswith('.csv'):
        csv_list.append(href)
```

```python
df_col['csv_links'] = pd.Series(csv_list)
df_col
```

### 4.[`10 pts`] Import the csv file for the dataset identified in #3 using the full link listed in the "`csv_links`" column for this dataset.
```python
csv_link = df_title['csv_links'].item()  # Use the item method to return value(s), rather than pd object
csv_link
```

```python
df_voi = pd.read_csv(csv_link)

# Create the new variable and rename rownames to state.name
df_voi['violent_crime'] = df_voi['Assault'] + df_voi['Murder'] + df_voi['Rape']
df_voi.rename(columns={'rownames': 'state.name'}, inplace=True)

df_voi
```

### 10.[`10 pts`] Create a new data frame that includes two columns from your state data denoting state names and violent crimes. Spread the state names to 50 unique columns with a single row that includes the violent crime data per state. Print the first five columns of the new dataset.

### 11.[`10 pts`] Take the dataset from question 10 and return a dictionary with each key denoting a state and each value indicating the square root of the value for violent crimes.

```python
# Apply the sqrt function to the violent_crime row to create a new row with the square root values
df_wide.loc['sqrt_violent_crime'] = df_wide.loc['violent_crime'].apply(np.sqrt)

# Convert to a dictionary
result_dict = df_wide.loc['sqrt_violent_crime'].to_dict()

print(result_dict)
```
